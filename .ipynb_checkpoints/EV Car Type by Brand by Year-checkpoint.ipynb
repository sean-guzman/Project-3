```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from CSV
car_brands = 'EVs for each Car Brand by Year.xlsx'
car_brands_df = pd.read_excel(car_brands)

car_brands_df.head()
```

```python
car_brands_df.replace('', np.nan, inplace=True)
car_brands_df
```

```python
car_brands_df = car_brands_df.fillna(0)
car_brands_df.head()
```

```python
for i in car_brands_df.columns:
    try:
        car_brands_df[[i]] = car_brands_df[[i]].astype(int)
    except:
        pass

car_brands_df.head()
```

```python
car_brands_df.drop(["Total"], axis = 1)
car_brands_df.head()
```
