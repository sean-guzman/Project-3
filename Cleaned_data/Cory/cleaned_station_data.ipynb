```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import pycountry
from pymongo import MongoClient
```

```python
df = pd.read_csv('us_stations.csv')
```

```python
df
```

```python
# Create a new DataFrame with selected columns
location_columns = ['AddressInfo,Country,Title', 'AddressInfo,Latitude', 'AddressInfo,Longitude','DateCreated','AddressInfo,Title','UsageType,Title']
location_df = df[location_columns].copy()
location_df
```

```python
df = location_df.rename(columns={
    'AddressInfo,Country,Title': 'Country',
    'AddressInfo,Latitude': 'Latitude',
    'AddressInfo,Longitude': 'Longitude',
    'DateCreated': 'Date_Created',
    'AddressInfo,Title': 'Title',
    'UsageType,Title': 'Usage_Type'
})

df
```

```python
# Load the shapefile data for states
states = gpd.read_file('tl_rd22_us_state/tl_rd22_us_state.shp')
```

```python
# Load the county shapefile
counties = gpd.read_file('tl_rd22_us_county/tl_rd22_us_county.shp')
```

```python
# Create a new column in the DataFrame to store the state and county information
df['State'] = ''
df['County'] = ''
```

```python
# Define the CRS of the latitude and longitude points
crs = "EPSG:4326"  # Assuming the coordinates are in WGS84
```

```python
# Convert latitude and longitude to Point geometries with the defined CRS
geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
geo_df = gpd.GeoDataFrame(df, geometry=geometry, crs=crs)
```

```python
# Reset the indices of the GeoDataFrames
states = states.reset_index(drop=True)
counties = counties.reset_index(drop=True)
```

```python
# Reproject the state and county shapefiles to match the CRS of the latitude and longitude points
states = states.to_crs(crs)
counties = counties.to_crs(crs)
```

```python
# Perform spatial join to assign state information to all points
joined_states = gpd.sjoin(geo_df, states, how='left', op='within')
```

```python
# Update the 'State' column in the DataFrame with the state information
df['State'] = joined_states['NAME']
```

```python
# Filter the joined DataFrame to keep only the rows with state information
joined_states_counties = joined_states.dropna(subset=['NAME'])
```

```python
# Perform spatial join to assign county information to the state-filtered points
joined_counties = gpd.sjoin(joined_states_counties, counties, how='left', op='within', lsuffix='_left', rsuffix='_right')
```

```python
# Update the 'County' column in the DataFrame with the county information
df['County'] = joined_counties['NAME__right']
```

```python
df
```

```python
df = df.fillna('NONE')
```

```python
df['Date_Created'] = pd.to_datetime(df['Date_Created']).dt.date
```

```python
desired_column_order = ['Title', 'Usage_Type','County', 'State', 'Country', 'Latitude', 'Longitude', 'Date_Created']
df = df.reindex(columns=desired_column_order)
```
