```python
import pandas as pd
import datetime as dt
```

```python
# Create year list for loop
year_list = ['2021', '2020', '2019', '2018', '2017', '2016']
```

```python
combined_vehicle_df = pd.DataFrame()

# Estasblish for loop
for i in year_list:
    str='https://afdc.energy.gov/vehicle-registration?year='
    final_str = str+ i
    
    # Read in HTML tables into a DataFrame
    df = pd.read_html(final_str)

    # Select table
    vehicle_df = df[0]
    
    # Drop Level
    vehicle_df.columns = vehicle_df.columns.droplevel()
    
    # Add column
    vehicle_df["Year"] = i
    
    # Concatanate the dataframes 
    combined_vehicle_df = pd.concat([vehicle_df, combined_vehicle_df])
```

```python
combined_vehicle_df
```

```python
ev_gas_df = combined_vehicle_df.copy()
```

```python
ev_gas_df.dtypes
```

```python
ev_gas_df["Electric_Hybrid_Vehicles_Total"] = ev_gas_df["Electric (EV)"] + ev_gas_df["Plug-In Hybrid Electric (PHEV)"] + ev_gas_df["Hybrid Electric (HEV)"]
```

```python
ev_gas_df = ev_gas_df.drop(columns = ['Biodiesel', 'Ethanol/Flex (E85)', 'Compressed Natural Gas (CNG)', 'Hydrogen',
                                     'Propane', 'Methanol', 'Diesel'], axis = 1)
```

```python
ev_gas_df['Total_Vehicles'] = ev_gas_df['Electric_Hybrid_Vehicles_Total'] + ev_gas_df['Gasoline']
ev_gas_df['EV_Hybrid_Ratio'] = ev_gas_df['Electric_Hybrid_Vehicles_Total'] / ev_gas_df['Total_Vehicles']
ev_gas_df
```

```python
ev_gas_df = ev_gas_df.rename(columns={'Electric (EV)': 'Electric_(EV)', 'Plug-In Hybrid Electric (PHEV)': 'Plug-In_Hybrid_Electric_(PHEV)',
                                     'Hybrid Electric (HEV)': 'Hybrid_Electric_(HEV)'})
ev_gas_df
```

```python
ev_gas_df.dtypes
```

```python
cleaned_df = ev_gas_df[['Year', 'State', 'Electric_(EV)', 'Plug-In_Hybrid_Electric_(PHEV)', 'Hybrid_Electric_(HEV)',
                        'Electric_Hybrid_Vehicles_Total', 'Gasoline', 'Total_Vehicles', 'EV_Hybrid_Ratio'
                      ]]
cleaned_df
```

```python
cleaned_df['Year'] = cleaned_df['Year'].astype('int')
```

```python
# Import cleaned CSV with the summary # of policies in effect, by year
policy_data = pd.read_csv('output/summary_policy_counts_bystate.csv')
policy_data= policy_data.rename(columns = {"Count": "Policy_Count"})
policy_data = policy_data.drop(columns='Unnamed: 0')
policy_data.head()
```

```python
# Merge policy data with cleaned vehicle data
merged_df = pd.merge(cleaned_df, policy_data, on = ["Year", 'State'], how = "left")
merged_df
```
