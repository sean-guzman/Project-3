```python
# Sort 'End Years' to visualize the end dates and if any policies ended from 2016-2021 
# Create a custom sort to deal with -inf values in 'End Year'
def custom_sort(x):
    if x == -np.inf:
        return float('inf')  # Place -inf at the end
    else:
        return x

sorted_end_years = policy_df['End Year'].apply(custom_sort).sort_values(ascending = False)
sorted_end_years.head(70)
```

```python
# CORRECT Function: Create function to iterate through data frame and count the number of incentives by year

summary_df = pd.DataFrame()
years = [2016, 2017, 2018, 2019, 2020, 2021]

for year in years:
    initiatives_count = ((recent_policy['Start Year'].astype(int) <= year) &
                         ((recent_policy['End Year'].astype(float) >= year) | (recent_policy['End Year'] == '-inf')))
    count_by_state = initiatives_count.groupby(recent_policy['State']).sum().reset_index()
    count_by_state['Year'] = year
    summary_df = pd.concat([summary_df, count_by_state])

summary_df = summary_df.rename(columns={0: 'Count'})
summary_df.head(50)
```
