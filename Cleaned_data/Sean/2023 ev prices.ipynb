# Average Prices of EVs versus New Cars

https://insideevs.com/news/565883/electric-car-prices-us/


```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit the site
url = 'https://insideevs.com/news/565883/electric-car-prices-us/'
browser.visit(url)
```

```python
# Parse the website
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Find the div containing table
table = html_soup.find('div', class_='table-wrapper').find('table')
```

```python
table
```

```python
data = []


for row in table.tbody.find_all('tr'):
    
    row_data = []
    
    for td in row.find_all('td'):
        
        row_data.append(td.text.strip())
    
    data.append(row_data)







```

```python
data[1]
```

```python
df = pd.DataFrame(data)
df.columns = df.iloc[0]

#remove first row from DataFrame
df = df[1:]
df
```

```python
# Quit the browsing session
browser.quit()
```

```python
df.dtypes
```

```python
df_cleaned = df
```

```python
df_cleaned
```

```python
new = df_cleaned["Model"].str.split(" ", n = 2, expand = True)
new
```

```python
df_cleaned ['Year'] = new[0]
df_cleaned ['Make'] = new[1]
df_cleaned ['Model'] = new[2]
```

```python
df_cleaned = df_cleaned[['Year', 'Make', 'Model', 'Base Price', 'Dest. Charge', 'Tax Credit', 'Effective Price']]

df_cleaned
```

```python
df_cleaned['Base Price'] = df_cleaned['Base Price'].str.replace('$', '')
df_cleaned['Base Price'] = df_cleaned['Base Price'].str.replace(',', '')
df_cleaned['Base Price'] = df_cleaned['Base Price'].str.replace('*', '')
```

```python
df_cleaned['Effective Price'] = df_cleaned['Effective Price'].str.replace('$', '')
df_cleaned['Effective Price'] = df_cleaned['Effective Price'].str.replace(',', '')
```

```python
df_cleaned['Dest. Charge'] = df_cleaned['Dest. Charge'].str.replace('+', '')
df_cleaned['Dest. Charge'] = df_cleaned['Dest. Charge'].str.replace('$', '')
df_cleaned['Dest. Charge'] = df_cleaned['Dest. Charge'].str.replace(',', '')
```

```python
df_cleaned['Tax Credit'] = df_cleaned['Tax Credit'].str.replace('$', '')
df_cleaned['Tax Credit'] = df_cleaned['Tax Credit'].str.replace(',', '')
df_cleaned['Tax Credit'] = df_cleaned['Tax Credit'].str.replace('N/A', '0')

```

```python
df_cleaned
```

```python
df_cleaned.drop(df_cleaned[df_cleaned['Base Price'] == ""].index, inplace = True)
```

```python
df_cleaned['Base Price'] = df_cleaned['Base Price'].astype(int)
df_cleaned['Dest. Charge'] = df_cleaned['Dest. Charge'].astype(int)
df_cleaned['Tax Credit'] = df_cleaned['Tax Credit'].astype(int)
df_cleaned['Effective Price'] = df_cleaned['Effective Price'].astype(int)
```

```python
df_cleaned.dtypes
```

```python
df_cleaned = df_cleaned.reset_index()
df_cleaned
```
