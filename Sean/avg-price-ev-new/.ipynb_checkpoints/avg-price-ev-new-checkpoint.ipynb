# Average Prices of EVs versus New Cars

https://caredge.com/guides/average-price-of-an-electric-car

```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
from datetime import date
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit https://caredge.com/guides/average-price-of-an-electric-car
url = 'https://caredge.com/guides/average-price-of-an-electric-car'
browser.visit(url)
```

```python
# Parse the website
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Find the table
table = html_soup.find('table')
```

```python
table
```

```python
df = pd.DataFrame(columns=['Dates', 'Average EV Price', 'New Car Average'])

df
    
```

```python
dates = []
avg_ev = []
avg_new = []

for row in table.thead.find_all('th'):
    dates.append(row.text.strip())

dates.pop(0)

avg_ev_costs = table.tbody.find('tr', class_="ninja_table_row_0 nt_row_id_470")
for row in avg_ev_costs.find_all('td'):
    avg_ev.append(row.text.strip())

avg_new_costs = table.tbody.find('tr', class_="ninja_table_row_1 nt_row_id_471")
for row in avg_new_costs.find_all('td'):
    avg_new.append(row.text.strip())
```

```python
df['Dates'] = dates
df['Average EV Price'] = avg_ev
df['New Car Average'] = avg_new

df
```

```python
# Quit the browsing session
browser.quit()
```
